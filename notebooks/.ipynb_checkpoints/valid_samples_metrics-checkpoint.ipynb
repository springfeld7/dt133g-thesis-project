# Valid Samples Metrics Analysis

This notebook explores statistical properties of the extracted dataset filtered to only include valid samples.
Valid samples are defined as those that:

- Belong to one of the following languages: Python, Java, or C++
- Have a label of either MACHINE_GENERATED or HUMAN_GENERATED
- Are successfully parsed by TranStructiver without any ERROR nodes in the resulting parse tree

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet("../data/valid_samples/valid_samples.parquet")
```

```python
metrics = [
    "char_count",
    "loc",
    "lloc",
    "identifier_density",
    "for_loop_density",
    "comment_density",
    "whitespace_ratio",
]
```

```python
print("=== DATASET OVERVIEW ===")
print(f"Total samples: {len(df)}\n")

print("=== LANGUAGE DISTRIBUTION ===")
print(df["language"].value_counts(), "\n")

print("=== LABEL DISTRIBUTION ===")
print(df["label"].value_counts(), "\n")

print("=== LANGUAGE × LABEL DISTRIBUTION ===")
print(df.groupby(["language", "label"]).size())
```

```python
label_means = df.groupby("label")[metrics].mean().T  # transpose so metrics are x-axis

ax = label_means.plot(
    kind="bar",
    figsize=(12, 5),
    width=0.8
)

plt.title("Average Metrics by Label")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.legend(title="label")

# add value labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.3f", fontsize=8, padding=2)

plt.tight_layout()
plt.show()
```

```python
lang_means = df.groupby("language")[metrics].mean()

lang_label_means = df.groupby(["language", "label"])[metrics].mean()

ax = lang_label_means.plot(kind="bar", figsize=(12, 6))

plt.yscale("log")
plt.title("Average Metrics per Language and Label (Log Scale)")
plt.ylabel("Value (log scale)")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")

ax.margins(y=0.2)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", fontsize=7, rotation=90, padding=2)

plt.tight_layout()
plt.show()
```

```python
METRIC = "lloc"

plt.figure(figsize=(12, 6))

sns.boxplot(
    data=df,
    x="language",
    y=METRIC,
    hue="label"
)

plt.yscale("log")
plt.title("lloc Distribution (Language × Label)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# --- SUMMARY TABLE ---
summary = (
    df.groupby(["language", "label"])[METRIC]
    .agg([
        ("count", "count"),
        ("mean", "mean"),
        ("median", "median"),
        ("std", "std"),
        ("p10", lambda x: x.quantile(0.10)),
        ("p25", lambda x: x.quantile(0.25)),
        ("p75", lambda x: x.quantile(0.75)),
        ("p90", lambda x: x.quantile(0.90)),
        ("p95", lambda x: x.quantile(0.95)),
        ("max", "max"),
    ])
)

print("\n=== SUMMARY TABLE (lloc) ===")
print(summary)
```

```python
METRIC = "char_count"

plt.figure(figsize=(12, 6))

sns.boxplot(
    data=df,
    x="language",
    y=METRIC,
    hue="label"
)

plt.yscale("log")
plt.title("char_count Distribution (Language × Label)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# --- SUMMARY TABLE ---
summary = (
    df.groupby(["language", "label"])[METRIC]
    .agg([
        ("count", "count"),
        ("mean", "mean"),
        ("median", "median"),
        ("std", "std"),
        ("p10", lambda x: x.quantile(0.10)),
        ("p25", lambda x: x.quantile(0.25)),
        ("p75", lambda x: x.quantile(0.75)),
        ("p90", lambda x: x.quantile(0.90)),
        ("p95", lambda x: x.quantile(0.95)),
        ("max", "max"),
    ])
)

print("\n=== SUMMARY TABLE (char_count) ===")
print(summary)
```

```python
LANGUAGES = ["C++", "Java", "Python"]
MIN_LLOC = 50
MIN_CHAR = 1000

# -----------------------------
# FILTER + COUNT
# -----------------------------

print("=== COUNTS AFTER FILTERING ===\n")

for lang in LANGUAGES:
    lang_df = df[df["language"] == lang]

    lloc_max = lang_df["lloc"].quantile(0.95)
    char_max = min(lang_df["char_count"].quantile(0.95), 6000)

    for label in ["HUMAN_GENERATED", "MACHINE_GENERATED"]:
        subset = lang_df[
            (lang_df["label"] == label) &
            (lang_df["lloc"] >= MIN_LLOC) &
            (lang_df["lloc"] <= lloc_max) &
            (lang_df["char_count"] >= MIN_CHAR) &
            (lang_df["char_count"] <= char_max)
        ]

        print(f"{lang} | {label}: {len(subset)}")

print("\nTarget per group: 1700")
```
